```python
import pandas as pd
```

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
display(file1.shape)
display(file2.shape)
display(file3.shape)
```

```python
display(file1.columns)
display(file2.columns)
display(file3.columns)
```

```python
cols1 = []
for colname in file1.columns:
    cols1.append(colname.lower())
file1.columns = cols1

cols2 = []
for colname in file2.columns:
    cols2.append(colname.lower())
file2.columns = cols2

cols3 = []
for colname in file3.columns:
    cols3.append(colname.lower())
file3.columns = cols3

display(file1.columns)
display(file2.columns)
display(file3.columns)
```

```python
file3 = file3.rename(columns={'state':'st'})
```

```python
file2 = file2[['customer', 'st', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
file3 = file3[['customer', 'st', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
```

```python
data = pd.concat([file1, file2, file3], axis=0)
```
